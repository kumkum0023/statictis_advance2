Que-9 How do you calculate the Z-score, and what does it represent in hypothesis testing?
Ans- ### üî¢ **How to Calculate the Z-Score**

The **Z-score** is a standardized value that tells you how many **standard deviations** an observation or sample statistic is from the population mean.

#### ‚úÖ **Formula for Z-Score (Single Observation)**

$$
Z = \frac{X - \mu}{\sigma}
$$

Where:

* $X$ = individual data point
* $\mu$ = population mean
* $\sigma$ = population standard deviation

---

#### ‚úÖ **Formula for Z-Score (Sample Mean, for Hypothesis Testing)**

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $\mu$ = population mean under the null hypothesis
* $\sigma$ = population standard deviation
* $n$ = sample size

This is the version used in **Z-tests**.

---

### üß† **What Does the Z-Score Represent?**

* The Z-score tells you how far (in standard deviations) your data is from the **null hypothesis mean**.
* It allows you to use the **standard normal distribution (mean = 0, SD = 1)** to calculate probabilities and compare against critical values.

---

### üìä **Interpreting Z-Scores**

| **Z-Score** | **Interpretation**                       |

QUE-14  What is a confidence interval, and how is it used to interpret statistical results?
ANS -

A **confidence interval** (CI) is a range of values, derived from sample data, that is likely to contain the **true population parameter** (such as a mean, proportion, etc.) with a certain level of confidence. It provides an estimated range of values that is likely to include the population parameter based on the sample data.

For example, a **95% confidence interval** means that if you were to take many samples and calculate their confidence intervals, about **95% of those intervals would contain the true population parameter**.

---

### ‚úÖ **Key Components of a Confidence Interval**

1. **Point Estimate**: The sample statistic (e.g., sample mean, sample proportion).
2. **Margin of Error**: A value that accounts for variability in the sample, which is used to create the range of the confidence interval.
3. **Confidence Level**: The probability (often expressed as a percentage, such as 95% or 99%) that the interval contains the true population parameter.

---

### üß† **How Is a Confidence Interval Calculated?**

The general formula for a confidence interval for a population mean when the population standard deviation ($\sigma$) is known is:

$$
\text{CI} = \bar{X} \pm Z_{\alpha/2} \left(\frac{\sigma}{\sqrt{n}}\right)
$$

Where:

* $\bar{X}$ = sample mean
* $Z_{\alpha/2}$ = Z-value corresponding to the desired confidence level (e.g., 1.96 for 95% CI)
* $\sigma$ = population standard deviation (if known)
* $n$ = sample size

If the population standard deviation is **unknown**, and you're using a **T-test** (for small sample sizes), the formula adjusts slightly:

$$
\text{CI} = \bar{X} \pm t_{\alpha/2} \left(\frac{s}{\sqrt{n}}\right)
$$

Where:

* $s$ = sample standard deviation
* $t_{\alpha/2}$ = t-value from the t-distribution for the desired confidence level

---

### üîç **Interpreting a Confidence Interval**

1. **Range of Estimates**: The CI provides a **range** in which the true population parameter is likely to fall. For example, if the 95% confidence interval for a sample mean is \[45, 55], we can say that we are **95% confident** that the true population mean lies between 45 and 55.

2. **Confidence Level**: The **confidence level** (e.g., 95%) tells you how confident you can be that the interval includes the true parameter. If you have a 95% confidence level, it means that in 95% of repeated sampling, the intervals will capture the true population parameter.

---

### ‚öñÔ∏è **What Does a Confidence Interval Tell You?**

* **Precision of Estimates**: A narrow CI indicates that the sample mean is **precisely estimated**, while a wider CI suggests more **uncertainty**.
* **Significance**: If a **hypothesized value** (such as 0 or a specific population mean) lies **outside the CI**, it suggests that the sample provides evidence against that value. For example, if a 95% CI for a population mean is \[10, 20], and the hypothesized population mean is 30, we would reject the hypothesis that the population mean is 30.

QUE - 16 How is Bayes' Theorem used in statistics, and what is its significance?
ANS - ### üìò **What Is Bayes' Theorem in Statistics?**

**Bayes' Theorem** is a fundamental concept in probability theory that allows you to update the probability of an event based on new evidence. It relates current evidence to prior knowledge, providing a way to revise our beliefs in light of new data.

Bayes' Theorem is used in a variety of statistical applications, from hypothesis testing to machine learning and decision-making. It provides a formal way to combine **prior probabilities** (initial beliefs) with **likelihoods** (new data) to calculate **posterior probabilities** (updated beliefs).

---

### ‚úÖ **Bayes' Theorem Formula**

Bayes' Theorem is expressed as:

$$
P(A | B) = \frac{P(B | A) P(A)}{P(B)}
$$

Where:

* **$P(A | B)$** is the **posterior probability** ‚Äî the probability of event **A** happening given that **B** has occurred.
* **$P(B | A)$** is the **likelihood** ‚Äî the probability of observing **B** given that **A** is true.
* **$P(A)$** is the **prior probability** ‚Äî the initial probability of event **A** before seeing the new evidence.
* **$P(B)$** is the **marginal likelihood** ‚Äî the probability of observing the evidence **B**, which can be found using the law of total probability.

---

### üß† **How Does Bayes' Theorem Work?**

Bayes' Theorem allows you to **update your beliefs** about the probability of an event as new information or evidence is acquired.

1. **Prior Probability** ($P(A)$): Initially, we start with a belief about the probability of an event based on previous knowledge or data.
2. **Likelihood** ($P(B | A)$): We obtain new evidence (e.g., an observation or experiment), and the likelihood is how probable this new evidence is, assuming the event $A$ is true.
3. **Posterior Probability** ($P(A | B)$): Using Bayes' Theorem, we calculate the updated probability of event $A$ considering both the prior knowledge and the new evidence.



---

### üß© **Significance of Bayes' Theorem in Statistics**

1. **Updating Beliefs**: Bayes' Theorem allows you to **revise** your beliefs based on new data. It is particularly useful in situations where information evolves over time, such as in medical diagnostics, stock market analysis, or machine learning.

2. **Incorporating Prior Knowledge**: Unlike frequentist statistics, which only uses the data at hand, Bayes' Theorem allows for the **incorporation of prior knowledge** or **historical data** into the analysis. This is useful when the available sample size is small or when we have reliable information from previous studies or experiments.

3. **Probabilistic Inference**: It provides a framework for making **probabilistic inferences** about unknown quantities. For example, in hypothesis testing, Bayes' Theorem allows you to update the probability of a hypothesis being true based on the observed data, instead of relying on a fixed significance threshold (like p-values).

4. **Decision Making Under Uncertainty**: Bayes' Theorem helps in decision-making when there is uncertainty. By combining prior beliefs with evidence, it enables you to make informed decisions with more certainty, especially when there are multiple possible outcomes.

5. **Applications in Machine Learning**: In **machine learning**, particularly in **Bayesian networks** and **Bayesian inference**, Bayes' Theorem is used for tasks like **classification**, **regression**, and **model selection**, where updating probabilities is key to refining models over time.

QUE-20 What is an ANOVA test, and what are its assumptions?
ANS - ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if there are statistically significant differences between them. The primary goal of ANOVA is to assess whether any of the group means differ from each other, based on sample data.

ANOVA tests the null hypothesis that all group means are equal, and the alternative hypothesis that at least one group mean is different.

Basic Idea:
ANOVA works by comparing the variance within groups to the variance between groups. If the variance between the groups is significantly larger than the variance within the groups, it suggests that the group means are not all equal.

üéØ Assumptions of ANOVA
For the results of an ANOVA test to be valid, several assumptions need to be met:

Independence of Observations:

The observations in each group must be independent of each other. This means that the data points in one group should not influence the data points in another group.

Normality:

The data in each group should be approximately normally distributed. This assumption ensures that the F-statistic follows the correct distribution under the null hypothesis. If the sample sizes are large, the Central Limit Theorem can help mitigate this assumption.

Homogeneity of Variance (Homogeneity of Variances):

The variances of the groups being compared should be approximately equal. This assumption is important because the F-statistic is sensitive to differences in group variances. This is sometimes tested using Levene's Test or Bartlett‚Äôs Test for homogeneity of variances.

Random Sampling:

The data should come from random samples. This ensures that the results are generalizable and not biased by the method of data collection.

```python
#QUE-1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results.
#ANS -
from statsmodels.stats.weightstats import ztest as ztest

def perform_z_test(sample_data, population_mean, alpha=0.05):
  """
  Performs a one-sample Z-test to compare a sample mean to a known population mean.

  Args:
    sample_data: A list or array of sample data.
    population_mean: The known mean of the population.
    alpha: The significance level (default is 0.05).

  Returns:
    A tuple containing the z-statistic and the p-value.
  """
  # Perform the Z-test
  z_statistic, p_value = ztest(x1=sample_data, value=population_mean)

  # Interpret the results
  print(f"Z-statistic: {z_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  if p_value < alpha:
    print(f"At the {alpha} significance level, we reject the null hypothesis.")
    print("There is sufficient evidence to conclude that the sample mean is significantly different from the population mean.")
  else:
    print(f"At the {alpha} significance level, we fail to reject the null hypothesis.")
    print("There is not sufficient evidence to conclude that the sample mean is significantly different from the population mean.")

  return z_statistic, p_value

# Example usage:
# Assume we have a sample of test scores and a known population mean for those scores.
sample_scores = [75, 80, 85, 90, 95, 78, 88, 92, 83, 87]
known_population_mean = 80

perform_z_test(sample_scores, known_population_mean)
```

```python
 #QUE-4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# The previous code already includes the Z-test function and plotting.
# This cell focuses on performing the Z-test and then plotting the decision regions.

# Define sample data and population mean
sample_data_for_plot = np.random.normal(loc=70, scale=15, size=100)
population_mean_for_plot = 75
alpha_for_plot = 0.05 # Significance level for the two-tailed test

# Perform the Z-test
z_stat_plot, p_value_plot = ztest(x1=sample_data_for_plot, value=population_mean_for_plot)

print(f"Z-statistic for plot: {z_stat_plot:.4f}")
print(f"P-value for plot: {p_value_plot:.4f}")

# Plot the Z-distribution and decision regions
z_critical_lower_plot = norm.ppf(alpha_for_plot / 2)
z_critical_upper_plot = norm.ppf(1 - alpha_for_plot / 2)

x_plot = np.linspace(-4, 4, 1000)
y_plot = norm.pdf(x_plot)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='Standard Normal Distribution (Z-distribution)')

# Shade the rejection regions
plt.fill_between(x_plot, y_plot, where=(x_plot <= z_critical_lower_plot), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x_plot, y_plot, where=(x_plot >= z_critical_upper_plot), color='red', alpha=0.3)

# Shade the acceptance region
plt.fill_between(x_plot, y_plot, where=(x_plot > z_critical_lower_plot) & (x_plot < z_critical_upper_plot), color='green', alpha=0.3, label='Acceptance Region')

# Mark the calculated Z-statistic
plt.axvline(z_stat_plot, color='blue', linestyle='dashed', linewidth=2, label=f'Calculated Z-statistic: {z_stat_plot:.4f}')

plt.title('Two-Tailed Z-Test Decision Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# QUE-5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import matplotlib.pyplot as plt
import numpy as np
def plot_type1_type2_errors(mu_null, mu_alt, sigma, n, alpha):
  """
  Calculates and visualizes Type 1 and Type 2 errors for a one-tailed Z-test.

  Args:
    mu_null: Population mean under the null hypothesis.
    mu_alt: Population mean under the alternative hypothesis (for Type 2 error calculation).
    sigma: Population standard deviation.
    n: Sample size.
    alpha: Significance level (probability of Type 1 error).
  """
  # Standard error
  se = sigma / np.sqrt(n)

  # Calculate critical value for a right-tailed test (assuming Ha > mu_null)
  # You can adjust for a left-tailed test by using norm.ppf(alpha)
  z_critical = norm.ppf(1 - alpha)
  x_critical = mu_null + z_critical * se

  # Calculate beta (probability of Type 2 error) for a right-tailed test
  # Beta is the probability of failing to reject H0 when H1 is true
  # This is the area under the alternative distribution to the left of the critical value
  z_beta = (x_critical - mu_alt) / se
  beta = norm.cdf(z_beta)

  # Power of the test (1 - beta)
  power = 1 - beta

  print(f"Type 1 Error (alpha): {alpha:.4f}")
  print(f"Type 2 Error (beta): {beta:.4f}")
  print(f"Power (1 - beta): {power:.4f}")

  # Visualize the errors
  x = np.linspace(mu_null - 4 * se, mu_alt + 4 * se, 1000)

  plt.figure(figsize=(12, 7))

  # Plot the null distribution
  y_null = norm.pdf(x, mu_null, se)
  plt.plot(x, y_null, label=f'Null Hypothesis (H0: Œº = {mu_null})', color='blue')
  plt.fill_between(x, y_null, where=(x >= x_critical), color='red', alpha=0.3, label=f'Type 1 Error (Œ±={alpha:.2f})')

  # Plot the alternative distribution
  y_alt = norm.pdf(x, mu_alt, se)
  plt.plot(x, y_alt, label=f'Alternative Hypothesis (Ha: Œº = {mu_alt})', color='green')
  plt.fill_between(x, y_alt, where=(x < x_critical), color='orange', alpha=0.3, label=f'Type 2 Error (Œ≤={beta:.2f})')

  # Plot the critical value
  plt.axvline(x_critical, color='black', linestyle='dashed', linewidth=1.5, label=f'Critical Value ({x_critical:.2f})')

  plt.title('Type 1 and Type 2 Errors Visualization (Right-Tailed Test)')
  plt.xlabel('Sample Mean')
  plt.ylabel('Probability Density')
  plt.legend()
  plt.grid(True)
  plt.show()

# Example Usage:
# Assume we are testing if the mean score of a new teaching method (Ha) is greater than the standard method (H0).
# Population mean under H0: 75
# Population mean under Ha: 78 (assuming a true effect size)
# Population standard deviation: 10
# Sample size: 30
# Significance level: 0.05

mu_null_example = 75
mu_alt_example = 78
sigma_example = 10
n_example = 30
alpha_example = 0.05

plot_type1_type2_errors(mu_null_example, mu_alt_example, sigma_example, n_example, alpha_example)

# Another example with different parameters
mu_null_example_2 = 100
mu_alt_example_2 = 105
sigma_example_2 = 15
n_example_2 = 50
alpha_example_2 = 0.01

plot_type1_type2_errors(mu_null_example_2, mu_alt_example_2, sigma_example_2, n_example_2, alpha_example_2)

```

```python
# QUE-6: Write a Python program to perform an independent T-test and interpret the results.

from scipy.stats import ttest_ind
import numpy as np

def perform_independent_t_test(sample1_data, sample2_data, alpha=0.05):
  """
  Performs an independent samples T-test and interprets the results.

  Args:
    sample1_data: A list or array of data from the first sample.
    sample2_data: A list or array of data from the second sample.
    alpha: The significance level (default is 0.05).

  Returns:
    A tuple containing the t-statistic and the p-value.
  """
  # Perform the independent T-test
  # equal_var=True assumes equal variances between the two samples.
  # If variances are not assumed to be equal, set equal_var=False (Welch's T-test)
  t_statistic, p_value = ttest_ind(a=sample1_data, b=sample2_data, equal_var=True)

  # Interpret the results
  print(f"Independent Samples T-test Results:")
  print(f"T-statistic: {t_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  if p_value < alpha:
    print(f"At the {alpha} significance level, we reject the null hypothesis.")
    print("There is sufficient evidence to conclude that the means of the two samples are significantly different.")
  else:
    print(f"At the {alpha} significance level, we fail to reject the null hypothesis.")
    print("There is not sufficient evidence to conclude that the means of the two samples are significantly different.")

  return t_statistic, p_value

# Example Usage:
# Assume we have test scores from two different groups of students.
group1_scores = [85, 90, 78, 92, 88, 75, 80, 83, 87, 95]
group2_scores = [70, 75, 80, 85, 72, 78, 83, 88, 76, 81]

perform_independent_t_test(group1_scores, group2_scores)

# Another example with simulated data
simulated_group1 = np.random.normal(loc=60, scale=10, size=30)
simulated_group2 = np.random.normal(loc=65, scale=10, size=30)

print("\nAnother Example (Simulated Data):")
perform_independent_t_test(simulated_group1, simulated_group2, alpha=0.01)
```

```python
# QUE-7: Perform a paired sample T-test using Python and visualize the comparison results.
#ANS-
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt
import numpy as np

def perform_paired_t_test_and_visualize(before_data, after_data, alpha=0.05):
  """
  Performs a paired samples T-test, interprets the results, and visualizes the comparison.

  Args:
    before_data: A list or array of data from the 'before' condition.
    after_data: A list or array of data from the 'after' condition.
    alpha: The significance level (default is 0.05).
  """
  if len(before_data) != len(after_data):
    print("Error: The lengths of 'before' and 'after' data must be the same for a paired T-test.")
    return

  # Perform the paired T-test
  # The null hypothesis is that the mean difference between the pairs is zero.
  t_statistic, p_value = ttest_rel(a=before_data, b=after_data)

  # Interpret the results
  print(f"Paired Samples T-test Results:")
  print(f"T-statistic: {t_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  if p_value < alpha:
    print(f"At the {alpha} significance level, we reject the null hypothesis.")
    print("There is sufficient evidence to conclude that there is a significant difference between the 'before' and 'after' measurements.")
  else:
    print(f"At the {alpha} significance level, we fail to reject the null hypothesis.")
    print("There is not sufficient evidence to conclude that there is a significant difference between the 'before' and 'after' measurements.")

  # Visualize the results
  plt.figure(figsize=(10, 6))

  # Plot individual paired data points
  for i in range(len(before_data)):
    plt.plot([1, 2], [before_data[i], after_data[i]], marker='o', linestyle='-', color='gray', alpha=0.5)

  # Plot the average of the 'before' and 'after' groups
  avg_before = np.mean(before_data)
  avg_after = np.mean(after_data)
  plt.plot([1, 2], [avg_before, avg_after], marker='o', linestyle='--', color='blue', linewidth=2, label=f'Average (Before: {avg_before:.2f}, After: {avg_after:.2f})')

  plt.xticks([1, 2], ['Before', 'After'])
  plt.ylabel('Measurement Value')
  plt.title('Paired Samples Comparison')
  plt.legend()
  plt.grid(True)
  plt.show()

# Example Usage:
# Assume we measure the blood pressure of a group of patients before and after a treatment.
blood_pressure_before = [145, 150, 140, 155, 160, 148, 152, 158]
blood_pressure_after = [138, 142, 135, 148, 155, 140, 145, 150]

perform_paired_t_test_and_visualize(blood_pressure_before, blood_pressure_after)

# Another example with simulated data (e.g., performance scores before and after training)
scores_before_training = np.random.normal(loc=70, scale=5, size=20)
scores_after_training = scores_before_training + np.random.normal(loc=3, scale=2, size=20) # Simulate an increase

print("\nAnother Example (Simulated Training Scores):")
perform_paired_t_test_and_visualize(scores_before_training, scores_after_training, alpha=0.01)
```

```python
# QUE-9: Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
#ANS-
import numpy as np
from scipy.stats import norm, t

def calculate_confidence_interval_mean(data, confidence_level=0.95, assumed_pop_std=None):
  """
  Calculates the confidence interval for a sample mean.

  Args:
    data: A list or array of numerical data.
    confidence_level: The desired confidence level (e.g., 0.95 for 95%).
    assumed_pop_std: The known population standard deviation.
                     If None (default), the sample standard deviation is used,
                     and a t-distribution is used.

  Returns:
    A tuple containing the lower and upper bounds of the confidence interval.
    Returns None if the data is empty.
  """
  if not data:
    print("Error: Input data is empty.")
    return None

  sample_mean = np.mean(data)
  sample_size = len(data)

  alpha = 1 - confidence_level

  if assumed_pop_std is not None:
    # Use Z-distribution if population standard deviation is known
    print("Using Z-distribution (assumed population standard deviation is known)")
    standard_error = assumed_pop_std / np.sqrt(sample_size)
    z_critical = norm.ppf(1 - alpha / 2)
    margin_of_error = z_critical * standard_error
  else:
    # Use T-distribution if population standard deviation is unknown
    print("Using T-distribution (population standard deviation is unknown)")
    if sample_size < 2:
        print("Error: Sample size must be at least 2 for T-distribution.")
        return None
    sample_std = np.std(data, ddof=1) # Use sample standard deviation (ddof=1 for unbiased estimate)
    standard_error = sample_std / np.sqrt(sample_size)
    degrees_of_freedom = sample_size - 1
    t_critical = t.ppf(1 - alpha / 2, degrees_of_freedom)
    margin_of_error = t_critical * standard_error

  lower_bound = sample_mean - margin_of_error
  upper_bound = sample_mean + margin_of_error

  print(f"\nSample Mean: {sample_mean:.4f}")
  print(f"Sample Size: {sample_size}")
  print(f"Confidence Level: {confidence_level * 100}%")
  print(f"Margin of Error: {margin_of_error:.4f}")
  print(f"Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

  return lower_bound, upper_bound

def explain_confidence_interval_significance():
  """
  Explains the significance of a confidence interval.
  """
  print("\nSignificance of a Confidence Interval:")
  print("A confidence interval provides a range of values within which the true population parameter (like the mean) is likely to lie, based on sample data.")
  print("The confidence level (e.g., 95%) represents the probability that the interval, if constructed for many samples, would contain the true population parameter.")
  print("It quantifies the uncertainty associated with estimating a population parameter from a sample.")
  print("- A wider interval indicates more uncertainty in the estimate.")
  print("- A narrower interval indicates greater precision in the estimate.")
  print("Confidence intervals are crucial for making inferences about the population and for understanding the reliability of statistical results.")
  print("If a hypothesized value for the population mean falls outside the confidence interval, it provides evidence against that hypothesized value (at the given confidence level).")

# Example Usage:

# Example 1: Using sample data (population standard deviation unknown - T-distribution)
sample_data_scores = [75, 80, 85, 90, 95, 78, 88, 92, 83, 87, 79, 86, 91, 84, 89]
calculate_confidence_interval_mean(sample_data_scores, confidence_level=0.95)

# Example 2: Using sample data with a known population standard deviation (Z-distribution)
# This is less common in practice, but demonstrates the Z-distribution case.
sample_data_heights = [165, 170, 175, 168, 172, 178, 169, 171, 176, 173]
known_pop_std_heights = 5.0 # Assume we know the population standard deviation
calculate_confidence_interval_mean(sample_data_heights, confidence_level=0.90, assumed_pop_std=known_pop_std_heights)

# Explain the significance
explain_confidence_interval_significance()
```

```python
# QUE-10: Write a Python program to calculate the margin of error for a given confidence level using sample data.
#ANS-
import numpy as np
def calculate_margin_of_error(data, confidence_level=0.95, assumed_pop_std=None):
  """
  Calculates the margin of error for a sample mean.

  Args:
    data: A list or array of numerical data.
    confidence_level: The desired confidence level (e.g., 0.95 for 95%).
    assumed_pop_std: The known population standard deviation.
                     If None (default), the sample standard deviation is used,
                     and a t-distribution is used.

  Returns:
    The margin of error.
    Returns None if the data is empty or sample size is too small for t-distribution.
  """
  if not data:
    print("Error: Input data is empty.")
    return None

  sample_size = len(data)
  alpha = 1 - confidence_level

  if assumed_pop_std is not None:
    # Use Z-distribution if population standard deviation is known
    print("Calculating Margin of Error using Z-distribution...")
    standard_error = assumed_pop_std / np.sqrt(sample_size)
    z_critical = norm.ppf(1 - alpha / 2)
    margin_of_error = z_critical * standard_error
  else:
    # Use T-distribution if population standard deviation is unknown
    print("Calculating Margin of Error using T-distribution...")
    if sample_size < 2:
        print("Error: Sample size must be at least 2 for T-distribution.")
        return None
    sample_std = np.std(data, ddof=1) # Use sample standard deviation
    standard_error = sample_std / np.sqrt(sample_size)
    degrees_of_freedom = sample_size - 1
    t_critical = t.ppf(1 - alpha / 2, degrees_of_freedom)
    margin_of_error = t_critical * standard_error

  print(f"Sample Size: {sample_size}")
  print(f"Confidence Level: {confidence_level * 100}%")
  print(f"Margin of Error: {margin_of_error:.4f}")

  return margin_of_error

# Example Usage:

# Example 1: Calculate Margin of Error using sample data (population standard deviation unknown - T-distribution)
sample_data_values = [25, 28, 30, 22, 35, 27, 32, 29, 31, 26]
calculate_margin_of_error(sample_data_values, confidence_level=0.95)

# Example 2: Calculate Margin of Error with an assumed population standard deviation (Z-distribution)
sample_data_weights = [150, 155, 160, 148, 152, 165, 158, 162, 153, 159]
assumed_pop_std_weights = 8.0 # Assume known population standard deviation
calculate_margin_of_error(sample_data_weights, confidence_level=0.90, assumed_pop_std=assumed_pop_std_weights)
```

```python
# QUE-11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
#ANS-
def bayesian_inference_example(prior_prob_A, prob_B_given_A, prob_B_given_not_A):
  """
  Implements a simple Bayesian inference example using Bayes' Theorem.

  Args:
    prior_prob_A: The prior probability of event A (P(A)).
    prob_B_given_A: The probability of observing evidence B given that A is true (P(B|A)).
    prob_B_given_not_A: The probability of observing evidence B given that A is false (P(B|~A)).

  Returns:
    The posterior probability of A given B (P(A|B)).
  """
  # Calculate the prior probability of not A (P(~A))
  prior_prob_not_A = 1 - prior_prob_A

  # Calculate the probability of observing evidence B (P(B))
  # Using the law of total probability: P(B) = P(B|A) * P(A) + P(B|~A) * P(~A)
  prob_B = (prob_B_given_A * prior_prob_A) + (prob_B_given_not_A * prior_prob_not_A)

  # Avoid division by zero if P(B) is 0
  if prob_B == 0:
    print("Probability of evidence B is zero. Cannot perform Bayesian inference.")
    return None

  # Calculate the posterior probability of A given B (P(A|B)) using Bayes' Theorem
  # P(A|B) = [P(B|A) * P(A)] / P(B)
  posterior_prob_A_given_B = (prob_B_given_A * prior_prob_A) / prob_B

  print("Bayesian Inference Process:")
  print(f"  Prior Probability of A (P(A)): {prior_prob_A:.4f}")
  print(f"  Likelihood (P(B|A)): {prob_B_given_A:.4f}")
  print(f"  Prior Probability of not A (P(~A)): {prior_prob_not_A:.4f}")
  print(f"  Probability of B given not A (P(B|~A)): {prob_B_given_not_A:.4f}")
  print(f"  Probability of Evidence B (P(B)): {prob_B:.4f}")
  print(f"  Posterior Probability of A given B (P(A|B)): {posterior_prob_A_given_B:.4f}")

  return posterior_prob_A_given_B

# Example Usage:
# Scenario: A medical test for a rare disease.
# Event A: The person has the disease.
# Evidence B: The medical test is positive.

# Assume the following probabilities:
# Prior probability of the disease in the population (P(A))
prevalence = 0.01 # 1% of the population has the disease

# Probability of a positive test given the person has the disease (Sensitivity) (P(B|A))
sensitivity = 0.95 # The test is positive for 95% of people with the disease

# Probability of a positive test given the person does NOT have the disease (False Positive Rate) (P(B|~A))
false_positive_rate = 0.05 # The test is positive for 5% of people without the disease

print("Medical Test Scenario:")
posterior_probability = bayesian_inference_example(prior_prob_A=prevalence,
                                                 prob_B_given_A=sensitivity,
                                                 prob_B_given_not_A=false_positive_rate)

print("\nInterpretation of the result:")
if posterior_probability is not None:
  print(f"After a positive test result (Evidence B), the updated probability that the person has the disease (Event A) is {posterior_probability:.4f}.")
  print(f"This shows how new evidence (a positive test) updates our belief about the event (having the disease), compared to the initial prevalence ({prevalence:.4f}).")

print("\nAnother Example: Spam Filter")
# Event A: The email is spam.
# Evidence B: The email contains the word "free".

# Assume:
# Prior probability of an email being spam (P(A))
prior_spam_prob = 0.3 # 30% of emails are spam

# Probability of an email containing "free" given it is spam (P(B|A))
prob_free_given_spam = 0.8 # 80% of spam emails contain "free"

# Probability of an email containing "free" given it is NOT spam (P(B|~A))
prob_free_given_not_spam = 0.1 # 10% of non-spam emails contain "free"

print("\nSpam Filter Scenario:")
posterior_spam_prob = bayesian_inference_example(prior_prob_A=prior_spam_prob,
                                                prob_B_given_A=prob_free_given_spam,
                                                prob_B_given_not_A=prob_free_given_not_spam)

print("\nInterpretation of the result:")
if posterior_spam_prob is not None:
  print(f"Given the email contains the word 'free' (Evidence B), the updated probability that it is spam (Event A) is {posterior_spam_prob:.4f}.")
  print(f"The initial belief was that {prior_spam_prob:.4f} of emails are spam. Seeing the word 'free' increased this probability to {posterior_spam_prob:.4f}.")
```

```python
# QUE13: Write a Python program to calculate the expected frequencies for a Chi-square test based on observed DATA
# ANS-

import pandas as pd
def calculate_expected_frequencies(observed_data):
  """
  Calculates the expected frequencies for a Chi-square test based on observed data.

  Args:
    observed_data: A pandas DataFrame representing the contingency table
                   of observed frequencies.

  Returns:
    A pandas DataFrame representing the expected frequencies.
  """
  # Get row and column totals
  row_totals = observed_data.sum(axis=1)
  col_totals = observed_data.sum(axis=0)
  grand_total = observed_data.sum().sum()

  # Initialize a DataFrame for expected frequencies
  expected_data = pd.DataFrame(index=observed_data.index, columns=observed_data.columns)

  # Calculate expected frequency for each cell
  for i in observed_data.index:
    for j in observed_data.columns:
      expected_data.loc[i, j] = (row_totals[i] * col_totals[j]) / grand_total

  return expected_data

# Example Usage (using the observed data from the previous example):
data_example = {'Voting_Preference_Yes': [30, 20],
                'Voting_Preference_No': [20, 30]}
index_example = ['Male', 'Female']
observed_data_example = pd.DataFrame(data_example, index=index_example)

print("Observed Frequencies:")
print(observed_data_example)

expected_frequencies_calculated = calculate_expected_frequencies(observed_data_example)

print("\nCalculated Expected Frequencies:")
print(expected_frequencies_calculated)

# Verify with chi2_contingency (it also returns expected frequencies)
# chi2_statistic, p_value, dof, expected_frequencies_from_statsmodels = chi2_contingency(observed_data_example)
# print("\nExpected Frequencies from statsmodels (for verification):")
# print(pd.DataFrame(expected_frequencies_from_statsmodels, index=index_example, columns=observed_data_example.columns))

# Another example
data_example_2 = {'Category_A': [10, 15, 20],
                  'Category_B': [5, 20, 15]}
index_example_2 = ['Group_1', 'Group_2', 'Group_3']
observed_data_example_2 = pd.DataFrame(data_example_2, index=index_example_2)

print("\n--- Another Example ---")
print("Observed Frequencies:")
print(observed_data_example_2)

expected_frequencies_calculated_2 = calculate_expected_frequencies(observed_data_example_2)

print("\nCalculated Expected Frequencies:")
expected_frequencies_calculated_2
```

```python
# QUE 14: Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
#ANS-
from scipy.stats import chisquare

def perform_goodness_of_fit_test(observed_frequencies, expected_frequencies):
  """
  Performs a Chi-square goodness-of-fit test to compare observed data to an expected distribution.

  Args:
    observed_frequencies: A list or array of observed frequencies for each category.
    expected_frequencies: A list or array of expected frequencies for each category.

  Returns:
    A tuple containing the chi2 statistic and the p-value.
  """
  if len(observed_frequencies) != len(expected_frequencies):
    print("Error: The number of observed and expected frequencies must be the same.")
    return None, None

  # Perform the Chi-square goodness-of-fit test
  chi2_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

  # Interpret the results
  print(f"Chi-square Goodness-of-Fit Test Results:")
  print(f"Chi-square Statistic: {chi2_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  alpha = 0.05 # Significance level

  print("\nInterpretation:")
  if p_value < alpha:
    print(f"At the {alpha} significance level, we reject the null hypothesis.")
    print("There is sufficient evidence to conclude that the observed distribution is significantly different from the expected distribution.")
  else:
    print(f"At the {alpha} significance level, we fail to reject the null hypothesis.")
    print("There is not sufficient evidence to conclude that the observed distribution is significantly different from the expected distribution.")

  return chi2_statistic, p_value

# Example Usage:
# Assume we rolled a six-sided die 120 times and recorded the frequencies.
# We want to test if the die is fair (expected frequencies are equal for each face).

observed_rolls = [20, 22, 18, 25, 15, 20] # Observed frequencies for faces 1 to 6
total_rolls = sum(observed_rolls)
expected_rolls = [total_rolls / 6] * 6 # Expected frequencies for a fair die

print("Observed Frequencies:", observed_rolls)
print("Expected Frequencies (Fair Die):", expected_rolls)

perform_goodness_of_fit_test(observed_rolls, expected_rolls)

# Another Example:
# Assume we observed the number of customers arriving per hour at a store.
# We want to test if the arrivals follow a specific distribution (e.g., Poisson with a known rate).
# For simplicity, let's test against a uniform distribution here for demonstration.

observed_arrivals_per_hour = [10, 12, 8, 15, 9] # Observed frequencies for 5 different hours
total_arrivals = sum(observed_arrivals_per_hour)
expected_arrivals_uniform = [total_arrivals / len(observed_arrivals_per_hour)] * len(observed_arrivals_per_hour)

print("\n--- Another Example (Customer Arrivals) ---")
print("Observed Frequencies:", observed_arrivals_per_hour)
print("Expected Frequencies (Uniform Distribution):", expected_arrivals_uniform)

perform_goodness_of_fit_test(observed_arrivals_per_hour, expected_arrivals_uniform)
```
